## Example 3.1 Illustration of the convergence of the Fourier series and Gibbsâ€™ phenomenon

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def set_graph():
    plt.figure(figsize=(12, 6))
    plt.xlabel("Time (s)")
    plt.ylabel("x(t)")
```

```python
t = np.arange(0.0, 1.0, 0.01)
x, x_list = np.zeros_like(t), []
pi = np.pi
for n in range(1, 80, 2):
    x = x + (4 / pi) * (1 / n * np.sin(2 * pi * n * t))
    x_list.append(x)
```

```python
set_graph()
indexs = [1, 4, -1]
for i in indexs:
    plt.plot(t, x_list[i])
plt.legend([str(t) for t in indexs])
plt.show()
```

## 3.2 Fourier coefficients of a square wave

```python
import numpy as np
import matplotlib.pyplot as plt
```

### case 1: r = 3

```python
r = 3
c_n = []

for n in range(1, 10 * r):
    tmp1, tmp2 = 0, 0
    
    for k in range(1, 2 * r + 1, 2):
        tmp_odd = np.exp(-1j * (k / r) * n * np.pi)
        tmp1 += tmp_odd

    for k in range(2, 2 * (r-1) + 1, 2):
        tmp_even = -np.exp(-1j * (k / r) * n * np.pi)
        tmp2 += tmp_even
    
    tmp = -0.5 + tmp1 + tmp2 - (0.5 * np.exp(-1j * 2 * n * np.pi))
    c_n.append(1j * tmp / (np.pi * n))
    
t = np.arange(0.0, n / r, 1 / r)
plt.stem(t, np.abs(c_n))
plt.show()
```

### case 2: r = 7.5

```python
r = 7.5
r2 = int(np.ceil(r))
c_n = []

for n in range(1, int(10 * r)):
    tmp1, tmp2 = 0, 0
    
    for k in range(1, 2 * r2 + 1, 2):
        tmp_odd = np.exp(-1j * (k / r) * n * np.pi)
        tmp1 += tmp_odd

    for k in range(2, 2 * (r2-1) + 1, 2):
        tmp_even = -np.exp(-1j * (k / r) * n * np.pi)
        tmp2 += tmp_even
    
    tmp = -0.5 + tmp1 + tmp2 - (0.5 * np.exp(-1j * 2 * n * np.pi))
    c_n.append(1j * tmp / (np.pi * n))
    
t = np.arange(0.0, n / r, 1 / r)
plt.stem(t, np.abs(c_n))
plt.show()
```
