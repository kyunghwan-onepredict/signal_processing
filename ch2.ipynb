## Example 2.1 Synthesis of periodic signals and almost periodic signals 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

#### Case 1: Periodic signal

```python
def draw_graph(x_axis, y_axis):
    plt.figure(figsize=(12, 6))
    plt.plot(x_axis, y_axis)
    plt.xlabel("Time (s)")
    plt.ylabel("x(t)")
    plt.show()
```

```python
amp1, amp2 = 1, 1
theta1, theta2 = 0, 0
p1, p2 = 1.4, 1.5
pi = np.pi
t = np.arange(0.0, 30.0, 0.01)
```

```python
x = amp1 * np.sin(2 * pi * p1 * t + theta1)
x += amp2 * np.sin(2 * pi * p2 * t + theta2)
draw_graph(t, x)
```

#### Case 2: Almost periodic signal

```python
p1, p2 = np.sqrt(2), 1.5
```

```python
x = amp1 * np.sin(2 * pi * p1 * t + theta1)
x += amp2 * np.sin(2 * pi * p2 * t + theta2)
draw_graph(t, x)
```
